#Calling the Dividends

```python
dividends = yf.Ticker("aapl").dividends
dividends
```

```python
#Plotting Apple dividends
import matplotlib.pyplot as plt
data = dividends

#Graphing the data
plt.plot(data.index, data.values)
plt.xlabel('Year')
plt.ylabel('Dividends')
plt.title('Dividend on Apple Stock (1987 - 2024)')
plt.grid(True)
plt.show()
```

```python
#Isolating Total Revenue from the Income Statement Yearly
# Access the income statement data as a Pandas DataFrame.
income_statement = yf.Ticker("aapl").income_stmt

# Access 'Total Revenue' using the appropriate index label.
total_revenue = income_statement.loc['Total Revenue']

# Print or use total_revenue.
print(total_revenue)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = total_revenue

# Convert the index to DateTimeIndex if it's not already
data.index = pd.to_datetime(data.index)

# Extract the year from the index
years = data.index.year

# Convert total revenue in billions
revenue_in_billions = data.values / 1e9  # Convert to billions

# Plotting the data against the extracted years
plt.plot(years, revenue_in_billions)
plt.xlabel('Year')
plt.ylabel('Total Revenue (Billions)')
plt.title('Total Revenue Apple Stock (Yearly 2020 - 2024)')
plt.grid(True)
plt.show()
```

```python
#Transposing the Income Statement
import yfinance as yf

# Get the income statement data
income_statement = yf.Ticker("aapl").income_stmt

# Transpose the DataFrame
rearranged_income_statement = income_statement.transpose()

# Now 'rearranged_income_statement' has dates as rows and line items as columns
print(rearranged_income_statement)
```
