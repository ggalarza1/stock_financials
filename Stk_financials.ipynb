# Fetching stocks historicals
(Income Statement, Balance Sheet, Cashflows, Dividends and also the SEC Filings the yahoo API has)

#Calling the Income Statement (Yearly/Quaterly)

```python
#Calling Ticker Income Statement Quarterly Financials
import yfinance as yf

income_statement = yf.Ticker("aapl").quarterly_financials
income_statement
```

```python
#Calling Ticker Income Statement Yearly Financials
import yfinance as yf

income_statement = yf.Ticker("aapl").financials
income_statement
```

```python
# You can also call the Income statement using the income_stmt
income = yf.Ticker("aapl").income_stmt
income
```

#Getting the Balance Sheet (Yearly/Quaterly)

```python
#If you call balance sheet, It will give you the results yearly.
import yfinance as yf

balancesheet_statement = yf.Ticker("aapl").balance_sheet
balancesheet_statement
```

```python
#Calling the quarterly balance sheet
balancesheet_statement = yf.Ticker("aapl").quarterly_balance_sheet
balancesheet_statement
```

#Getting the Cashflow (Yearly/Quaterly)

```python
# Calling cashflow gives you the result yearly
cashflow = yf.Ticker("aapl").cashflow
cashflow
```

```python
# Calling cashflow gives you the result yearly
cashflow = yf.Ticker("aapl").quarterly_cashflow
cashflow
```

#Calling the SEC filings Yahoo API has and how they are using it
They use AWS

```python
import yfinance as yf

sec_filings_data = yf.Ticker("aapl").sec_filings
sec_filings_data
```

```python
#Isolating Total Revenue from the Income Statement Yearly
# Access the income statement data as a Pandas DataFrame.
income_statement = yf.Ticker("aapl").income_stmt

# Access 'Total Revenue' using the appropriate index label.
total_revenue = income_statement.loc['Total Revenue']

# Print or use total_revenue.
print(total_revenue)
```

```python
#Transposing the Income Statement
import yfinance as yf

# Get the income statement data
income_statement = yf.Ticker("aapl").income_stmt

# Transpose the DataFrame
rearranged_income_statement = income_statement.transpose()

# Now 'rearranged_income_statement' has dates as rows and line items as columns
print(rearranged_income_statement)
```
